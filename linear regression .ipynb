```python
df = df.drop([ "Unnamed: 0","property_id","location_id","page_url","date_added","agent","agency"],axis = 1)
```

```python
df
```

```python
df['property_type'].describe()
```

```python
df['property_type'].unique()
```

```python
df['property_type'] = df['property_type'].replace({'Flat':0, 'House':1, 'Penthouse':2, 'Farm House':3, 'Lower Portion':4,
                             'Upper Portion':5, 'Room':6})
```

```python
df['city'].unique()
```

```python
df['city'] = df['city'].replace({'Islamabad':0, 'Lahore':1, 'Faisalabad':2, 'Rawalpindi':3, 'Karachi':4})
```

```python
df = df.drop(['location','province_name'],axis = 1)
```

```python
df
```

```python
df['purpose'].unique()
```

```python
df['purpose'] = df['purpose'].replace({'For Sale':0, 'For Rent':1})
```

```python
df
```

```python
# as to train need input and output we divide
x = df.drop(['price'], axis = 1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
model = LinearRegression() 
```

```python
model.fit(X_train,y_train)
```

```python
df.info()
```

# how to find accuracy of the model 

```python
model.coef_   #line of best fit coords 
```

```python
model.feature_names_in_  #all data in 
```

as the data is found by a line of best fit, the %error in value is found by cacluating the distance of point from line 

```python
# to use the test data to train the model

Y_Prediction = model.predict(X_test)
```

```python
# to find difference between the model output and true output 

from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,Y_Prediction)
```
