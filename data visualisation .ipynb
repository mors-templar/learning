subplot means that 1 graph has multiple other graphs 

```python
plt.subplot(221) # first 2 = number of rows , second 2 = number of ccolumns, 1 = place on graph (comes first , 2nd or etc)
plt.plot(x,y)
plt.title('plot 1')

plt.subplot(222) 
plt.plot(x,y)
plt.title('plot 2')

plt.subplot(223)
plt.plot(x,y)
plt.title('plot 3')

plt.subplot(224) 
plt.plot(x,y)
plt.title('plot 4')
```
