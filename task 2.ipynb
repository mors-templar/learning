```python
df_15_1 = file.select('Title')
```

```python
rdd=df_15_1.rdd
```

```python
rdd.collect()
```

```python
long_titles=rdd.filter(lambda x:int(len(x[0])) > 50).collect()
```

```python
rddColumns = ["Title"] #name of the column 



#### convert to df #############################################################
df_15_2 = spark.createDataFrame(long_titles, schema = rddColumns)
# dataframe = function(rdd to convert , schema = columns as an array)
df_15_2.show(truncate=False)
    
```
